```python
import seaborn as sns
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import scienceplots
plt.style.use('ieee')
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.labelweight'] = 'bold'
```

```python
labels_1 = ['Prompt1', 'Prompt2', 'Prompt3', 'Prompt4', 'Random Forest', 'MatTen', 'Darwin']
mae_bulk = [9.04, 8.12, 8.60, 7.74, 10.58, 7.98, 11.35]
r2_bulk = [0.947, 0.955, 0.954, 0.963, 0.927, 0.959, 0.918]
mae_cij = [2.79, 2.45, 2.67, 2.32, 3.33, 2.43, 3.47]
r2_cij = [0.947, 0.957, 0.956, 0.965, 0.924, 0.963, 0.915]
```

```python
plt.gca().invert_xaxis()
markers = ['o', 'o','o', '*', 'o', 'o', 'o']
colors = ['#669bbc', '#0077b6', '#03045e', '#c1121f', '#2d6a4f', "#74c69d", "#c9ada7"]
for mae, r2, label, marker, color in zip(mae_bulk, r2_bulk, labels_1, markers, colors):
    if label == 'Prompt4':
        plt.scatter(mae, r2, label=label, marker=marker, s=150, color=color)  # Increased size to 150
        plt.text(mae-0.2, r2 + 0.006, 'Prompt4', ha='center', va='top', fontsize=8)
    elif label == 'Prompt3':
        plt.scatter(mae, r2, label=label, marker=marker, s=50, color=color)  # Increased size to 50
        plt.text(mae+0.55, r2, label, ha='center', va='top', fontsize=8)
    elif label == 'MatTen':
        plt.scatter(mae, r2, label=label, marker=marker, s=50, color=color)  # Increased size to 50
        plt.text(mae+0.5, r2+0.0015, label, ha='center', va='top', fontsize=8)
    elif label == 'Prompt2':
        plt.scatter(mae, r2, label=label, marker=marker, s=50, color=color)
        plt.text(mae-0.53, r2-0.001, label, ha='center', va='top', fontsize=8)
    else:
        plt.scatter(mae, r2, label=label, marker=marker, s=50, color=color)  # Increased size to 50
        plt.text(mae, r2-0.0015, label, ha='center', va='top', fontsize=8)
plt.xlim(12, 7.0)
plt.ylim(0.91, 0.97)
plt.xlabel('MAE (GPa)', fontsize=9)
plt.ylabel('$R^2$', fontsize=9)
plt.tick_params(axis='both', labelsize=9)  # Set the font size for both x and y axis ticks
plt.savefig("different_methods_bulk.pdf", dpi=600, bbox_inches='tight')
plt.show()
```
